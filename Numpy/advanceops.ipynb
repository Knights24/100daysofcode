```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from traceback import print_tb


sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])



print("==== DATA ====")
print("\nShape Of data", sales_data.shape)
print("\n Filtering data    ", sales_data[:3])
print("\n Filtering data    ", sales_data[:, 1 : ]) # values after first colm 
print(" \n Sum of clm ", np.sum(sales_data[: , 1: ], axis= 0) )
print("\n min sales row ",np.min(sales_data[:,1: ], axis= 1))
print("\n max sales col ",np.max(sales_data[:,1: ], axis= 0))
print("\n mean sales row ",np.mean(sales_data[:,1: ], axis= 1))

cumsum = np.cumsum(sales_data[: , 1: ],axis= 1)
print("\n cummulative sales row ",cumsum)


```

```python
plt.figure(figsize= (6, 6))
plt.plot(np.mean(cumsum , axis= 0))
plt.title(" Avg cummulative sales")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
